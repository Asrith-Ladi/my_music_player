### ‚úÖ Step-by-step Instructions:

#### 1. **Navigate to your project folder:**

```bash
cd path/to/your/project
```

> Replace `path/to/your/project` with your actual folder path.

---

#### 2. **Create the conda environment inside the project folder:**

```bash
conda create --prefix ./env python=3.11
```

> * `--prefix ./env`: Tells conda to create the environment in the `env` subfolder of the current directory.
> * You can name the folder anything (like `./.venv` or `./myenv`) and change the Python version as needed.

---

#### 3. **Activate the environment:**

```bash
conda activate ./env
```

> ‚ö†Ô∏è Use the **relative path** to activate prefix-based environments.
